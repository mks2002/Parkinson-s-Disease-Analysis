```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### **0 - Healthy , 1 - Parkinson's disease**

```python
# Remove the column named 'name'
parkinsons_data = parkinsons_data.drop(columns=['name'])
```

```python
# number of rows and columns in the dataframe
parkinsons_data.shape
```

## **Data Visualization**

```python
sns.set()
```

```python
# Number of columns per row
columns_per_row = 4

# Number of rows needed
num_columns = len(parkinsons_data.columns)
num_rows = (num_columns // columns_per_row) + int(num_columns % columns_per_row != 0)

# Create subplots with increased height
fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(14,14))  # Increased height from 16 to 24

# Flatten axes for easy iteration
axes = axes.flatten()

# Create a distribution plot for each column
for i, column in enumerate(parkinsons_data.columns):
    sns.histplot(data=parkinsons_data, x=column, ax=axes[i])
    axes[i].set_title(column)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

## **X, Y, Train & Test Split**

```python
X = parkinsons_data.drop(columns=['status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

## **Standardizing the data**

```python
scaler = StandardScaler()
```

```python
x_train_std = scaler.fit_transform(X_train)
x_test_std = scaler.transform(X_test)
```

```python
print(x_train_std.shape)
print(x_test_std.shape)
```

```python
print(x_train_std)
print(x_test_std)
```

## **Model Training & Evaluation**

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(x_train_std, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(x_train_std)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(x_test_std)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

## **Build a Predictive System**

```python
input_data = X_test.iloc[25]

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
```

```python
print(prediction)

if (prediction[0] == 0):
  print("The Person does not has Parkinsons Disease .")
else:
  print("The Person has Parkinsons Disease .")
```
